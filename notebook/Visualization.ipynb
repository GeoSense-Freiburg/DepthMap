# Result of depth maps

```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from PIL import Image
from IPython.display import Video

# Get the working directory of this file
working_dir = Path().resolve()
root_dir = working_dir.parent
print(f"Project directory: {root_dir}")
```

### Example 1

```python
rgb_img = root_dir / "test/images/G5Bullet_56_2024-10-09_13_18_00_calibrated.jpg"
depth_map = root_dir / "test/output/depth_map/images/G5Bullet_56_2024-10-09_13_18_00_calibrated_depth.jpg"
depth_arr = np.load(root_dir / "test/output/depth_map/arrays/G5Bullet_56_2024-10-09_13_18_00_calibrated_depth.npy")
```

```python
save = True

# Read images
rgb = Image.open(rgb_img)
depth = Image.open(depth_map)

# Create figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot RGB image
ax1.imshow(rgb)
ax1.set_title('RGB Image')
ax1.axis('off')

# Plot depth map
ax2.imshow(depth, cmap='viridis')
ax2.set_title('Depth Map')
ax2.axis('off')

plt.tight_layout()

if save:
    plt.savefig(root_dir / "test/visualization/images/rgb_depth_map_example1.png")

plt.show()
```

```python
print(f"Minimum distance: {depth_arr.min():.2f}m")
print(f"Maximum distance: {depth_arr.max():.2f}m")
```

### Example 2

```python
rgb_img = root_dir / "test/images/G5Bullet_55_2024-10-08_12_42_00_calibrated.jpg"
depth_map = root_dir / "test/output/depth_map/images/G5Bullet_55_2024-10-08_12_42_00_calibrated_depth.jpg"
depth_arr = np.load(root_dir / "test/output/depth_map/arrays/G5Bullet_55_2024-10-08_12_42_00_calibrated_depth.npy")
```

```python
# Read images
rgb = Image.open(rgb_img)
depth = Image.open(depth_map)
depth_array = np.array(depth)

# Convert depth to single channel (read as 3 channels)
depth_array = np.array(depth)
if len(depth_array.shape) > 2:
    depth_array = depth_array[:,:,0]  # take first channel only

# Create figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot RGB image
ax1.imshow(rgb)
ax1.set_title('RGB Image')
ax1.axis('off')

# Plot depth map
ax2.imshow(depth_array, cmap='viridis_r', vmin=np.percentile(depth_array, 2), vmax=np.percentile(depth_array, 98))
ax2.set_title('Depth Map')
ax2.axis('off')

plt.tight_layout()

if save:
    plt.savefig(root_dir / "test/visualization/images/rgb_depth_map_example2.png")

plt.show()
```

```python
print(f"Minimum distance: {depth_arr.min():.2f}m")
print(f"Maximum distance: {depth_arr.max():.2f}m")
```
